```python
class ArrayStack:
    def __init__(self, size=None):
        if size is None:
            self.size = 10
        self.size = size
        self.data = [None] * self.size
        self.top = -1

    def palindrome(self):
        copy = self.data[:self.top+1].copy()
        normal = "".join(str(copy[j]) for j in range(len(copy)))
        reverse = "".join(str(copy.pop()) for j in range(len(copy)))

        if normal == reverse:
            print("The string is a palindrome")
        else:
            print("The string is not a palindrome")

    def isEmpty(self):
        return self.top == -1

    def isFull(self):
        return self.top == self.size - 1

    def peek(self):
        if self.isEmpty():
            raise Exception("Stack is empty")
        return self.data[self.top]

    def push(self, data):
        if self.isFull():
            raise Exception("Stack is full")
        self.top += 1
        self.data[self.top] = data

    def pop(self):
        if self.isEmpty():
            raise Exception("Stack is empty")
        popped_data = self.data[self.top]
        self.data[self.top] = None
        self.top -= 1
        return popped_data

    def printValues(self):
        if self.isEmpty():
            raise Exception("Stack is empty")
        iterator = 0
        while True:
            if self.data[iterator] is None:
                break
            print(self.data[iterator])
            if iterator == self.size - 1:
                break
            iterator += 1


# Stack of size 10
stack = ArrayStack(10)
# Pushing 8 elements
for i in range(1, 9):
    stack.push(i)
# Checking the top most value
print(stack.peek())
# Popping 3 values --> rem 5
for i in range(1, 4):
    print(f"{i}th pop = {stack.pop()}")
# Printing stack contents
stack.printValues()
# Trying to push 6 elements --> an exception will be raised
for i in range(10, 17):
    stack.push(i)
# Trying to print stack contents
stack.printValues()
```
