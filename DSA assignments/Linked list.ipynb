```python
class Node:
    def __init__(self, data=None, next=None):
        self.next = next
        self.data = data


class LinkedList:
    def __init__(self):
        self.head = None

    def insertBeginning(self, data):
        self.head = Node(data, self.head)

    def insertEnd(self, data):
        if self.head is None:
            self.head = Node(data, None)
        else:
            iterator = self.head
            while iterator.next:
                iterator = iterator.next
            iterator.next = Node(data, None)

    def printList(self):
        if self.head is None:
            raise Exception("The list is empty")
        else:
            iterator = self.head
            print("Linked list = ", end="")
            while iterator:
                print(str(iterator.data) + "-->", end="")
                iterator = iterator.next
            print("None")

    def getLength(self):
        if self.head is None:
            return 0
        else:
            iterator = self.head
            count = 0
            while iterator:
                count += 1
                iterator = iterator.next
            return count

    def insertSpecificIndex(self, data, index):
        if index < 0 or index >= self.getLength():
            raise Exception("Invalid index")
        elif index == 0:
            self.insertBeginning(data)
        elif index == self.getLength():
            self.insertEnd(data)
        else:
            iterator = self.head
            count = 0
            while iterator:
                if count == index - 1:
                    iterator.next = Node(data, iterator.next)
                    break
                iterator = iterator.next
                count += 1

    def removeAtIndex(self, index):
        if index < 0 or index >= self.getLength():
            raise Exception("Invalid index")
        elif index == 0:
            self.head = self.head.next
        else:
            count = 0
            iterator = self.head
            while iterator:
                if count == index - 1:
                    iterator.next = iterator.next.next
                    break
                iterator = iterator.next
                count += 1

# Instantiation
linked_list = LinkedList()
# Adding the months inorder
months = ["jan", "feb", "march", "may", "june", "july"]
for month in months:
    linked_list.insertEnd(month)
linked_list.printList()
# inserting at the third index
linked_list.insertSpecificIndex("april", 3)
linked_list.printList()
linked_list.removeAtIndex(6)
# printing the final list
linked_list.printList()
```
