```python
class HashTable:
    def __init__(self):
        # Setting the initial capacity in the ram
        self.MAX = 10
        # Initializing empty linked lists in the specific memory locations
        self.array = [[] for i in range(self.MAX)]

    """
    Getting the hash value of a string - modulo(string index, capacity)
    The hash value serves as an index to transverse through the first layer list
    """

    def getHash(self, key):
        sum = 0
        for character in key:
            # ord - function to get the ascii value
            sum += ord(character)
        # calculating hash value using the sum of ascii values
        return sum % self.MAX

    # This function sets an item in the data structure: instantiation --> t[string index] = value
    def __setitem__(self, key, value):
        # Getting the integer index using a hash function
        h = self.getHash(key)
        # Check to see if it is an update to an existing key or a new full on tuple
        found = False
        # Transversing through the linked list
        for index, element in enumerate(self.array[h]):
            # Seeing if there are two values and the date value is equal to the key
            if len(element) == 2 and element[0] == key:
                # Updating the initial value
                self.array[h][index] = (key, value)
                # Changing boolean state of found
                found = True

        # If found is false means that it is not update but a full insert
        if not found:
            self.array[h].append((key, value))

    # This is a getter method - gets the value of the string index provided: instantiation --> print(t[string index])
    def __getitem__(self, key):
        # string to integer index conversion
        h = self.getHash(key)
        # Indicator whether the key is found or not
        found = False
        # linked list transversal
        for index, element in enumerate(self.array[h]):
            # if the string index = date set the found to true and return the value
            if element[0] == key:
                found = True
                return element[1]
        # if not found raise an error
        if not found:
            raise Exception("Invalid date")

    # deletion : instantiation --> del t[string index]
    def __delitem__(self, key):
        # getting integer index
        h = self.getHash(key)
        # checking if the key is found
        found = False
        # list transversal
        for index, element in enumerate(self.array[h]):
            # finding the location of the key
            if element[0] == key:
                found = True
                # deletion
                del self.array[h][index]
        if not found:
            raise Exception("Invalid date")


t = HashTable()
t["march 6"] = 120
t["march 8"] = 67
t["march 9"] = 4
t["march 17"] = 459

print(t["march 17"])
```
